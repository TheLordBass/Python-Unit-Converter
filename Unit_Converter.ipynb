```python
convert_from = input("Enter staring unit of measurement(inches, feet, yards)")

# Converting from Inches
convert_to = input("Enter unit of measurement to convert to(inches, feet, yards)")
if convert_from.lower() in ['inches','in','inch']:
    number_of_inches = int(input("Enter number of inches to convert"))
    if convert_to.lower() in ['feet','ft', 'foot']:
        print(number_of_inches, 'in','=',number_of_inches * 0.0833333,'ft')
    elif convert_to.lower() in ['yards','yd', 'yards']:
        print(number_of_inches, 'in','=',number_of_inches * 0.0277778,'yd')
    else: 
        print('Wrong input')

# Converting from Feet
elif convert_from.lower() in ['feet','ft', 'foot']:
    number_of_feet = int(input("Enter number of feet to convert"))
    if convert_to.lower() in ['inches','in','inch']:
        print(number_of_feet, 'ft','=',number_of_feet * 12,'in')
    elif  convert_to.lower() in ['yards','yd', 'yards'] :
        print(number_of_feet, 'ft','=',number_of_feet * 0.333333,'yd')
    else: 
        print('Wrong Input')

# Converting from Yards
elif convert_from.lower() in ['yards','yd', 'yards']:
    number_of_yards = int(input("Enter number of yards to convert"))
    if convert_to.lower() in ['inches','in','inch']:
        print(number_of_yards, 'yd','=',number_of_yards * 36,'in')
    elif convert_to.lower() in ['feet','ft', 'foot']:
        print(number_of_yards, 'yd','=',number_of_yards * 3,'ft')
    else: 
        print('Wrong Input')
else:
    print('Please enter Inches, Feet or Yards')
    
```
